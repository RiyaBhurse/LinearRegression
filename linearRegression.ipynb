We will choose a line which gives the least cost function!
The eq~ of a line is y=mx+b
to find price we will write it as:
price = m*area+b

m=>slope
b=>intercept
area=>independent variable
price=>dependent variable


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
df.size
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

by now the model has found the value of coefficent and intercept.

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

```python
d = pd.read_csv("areas.csv")
d
```

```python
prices = reg.predict(d)
prices
```

```python
d['price'] = prices
d
```

```python
d.to_csv("predicted_prices.csv", index=False)
```
